```python
# Ball Track
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import math

width = 600
height = 600
ball_pos = []
ball_vel = []
ball_radius = 20
inner_track_radius = 150
outer_track_radius = 280
laps = 0
score = 0
life = 6
over_half = False

def distance(pos1, pos2):
    return math.sqrt((pos1[0] - pos2[0]) ** 2 + (pos1[1] - pos2[1]) ** 2)

def draw(canvas):
    global over_half, laps, score, ball_pos, ball_vel
    mid = [width / 2, height / 2]
    
    # Track
    canvas.draw_circle(mid, outer_track_radius, 1, "White", "Black")
    canvas.draw_circle(mid, inner_track_radius, 1, "White", "Green")
    
    # lines
    canvas.draw_line((mid[0], mid[1] - inner_track_radius), (mid[0], mid[1] - outer_track_radius), 5, "White")
    canvas.draw_line((mid[0], mid[1] + inner_track_radius), (mid[0], mid[1] + outer_track_radius), 1, "White")
    
    # Ball
    ball_pos[0] += ball_vel[0]
    ball_pos[1] += ball_vel[1]
    canvas.draw_circle(ball_pos, ball_radius, 1, "Red", "Red")
    
    # Score
    canvas.draw_text("Laps:", [265, 230], 30, "Aqua")
    canvas.draw_text(str(laps), [295, 280], 30, "Aqua")
    canvas.draw_text("Score:", [260, 340], 30, "Aqua")
    canvas.draw_text(str(score), [295, 390], 30, "Aqua")
    canvas.draw_text("lifes = " + str(life), [460, 50], 30, "Aqua")
    
    
    if distance(ball_pos, mid) > outer_track_radius - ball_radius or distance(ball_pos, mid) < inner_track_radius + ball_radius:
        new_game()
    elif over_half and abs(ball_pos[0] - mid[0]) < ball_radius and ball_pos[1] < mid[1]:
        over_half = False
        laps += 1
        if laps > score:
            score = laps
    elif abs(ball_pos[0] - mid[0]) < ball_radius and ball_pos[1] > mid[1]:
        over_half = True
        
    if life == 0:
        ball_pos = [-20,-20]
        ball_vel = [0,0]
        canvas.draw_text("Game Over", [180, 310], 50, "Rad")
        
def new_game():
    global ball_pos, ball_vel, num_laps, over_half, life, score
    ball_pos = [width / 2, height / 2 - (outer_track_radius + inner_track_radius) / 2]
    ball_vel = [0, 0]
    laps = 0
    score = 0
    over_half = False
    if life > 0:
        life -= 1
        music = simplegui.load_sound("file:///C:/Users/PC-03/Desktop/music.ogg.ogg")
        
def reset():
    global ball_pos, ball_vel, laps, over_half, life, score
    ball_pos = [width / 2, height / 2 - (outer_track_radius + inner_track_radius) / 2]
    ball_vel = [0, 0]
    laps = 0
    score = 0
    life = 5
    over_half = False
    
def keydown_handler(key):
    acc = 1
    if key==simplegui.KEY_MAP["left"]:
        ball_vel[0] -= acc
    elif key==simplegui.KEY_MAP["right"]:
        ball_vel[0] += acc
    elif key==simplegui.KEY_MAP["down"]:
        ball_vel[1] += acc
    elif key==simplegui.KEY_MAP["up"]:
        ball_vel[1] -= acc
        

# Frame
frame = simplegui.create_frame("Ball Track", width, height) 

# Event Handlers
frame.set_draw_handler(draw)
frame.set_keydown_handler(keydown_handler)
frame.set_canvas_background("Green")
frame.add_button("Restart", reset)

# Start
new_game()
frame.start()
```
