# wk-06_04  Blackjack

블랙잭 규칙

1) 딜러를 포함하여 여러명이 게임을 한다.  
2) 카드의 숫자 계산은 카드에 써있는 숫자 그대로, J, Q, K는 10에 해당하며, A(에이스)는 1 혹은 11 어느 쪽으로도 계산할 수 있다.   
3) 카드는 52장(조커 제외)을 사용한다.  
4) 딜러의 첫 번째 카드를 엎어 놓은 것 이외에는 모든 참가자의 카드가 공개된다.    
5) 카드 두 장을 기본적으로 지급받게 되며, 카드 숫자를 합쳐 가능한 21을 넘지 않으면서 21에 가깝게 만들면 이기는 게임.     
6) 처음 받은 2장 합쳐 21이 나오는 경우 블랙잭이 되며, 21이 되지 않았을 경우 원한다면 얼마든지 카드를 계속 뽑을 수 있다.   
7) 21을 초과하게(22 이상) 되었을 경우에는 '버스트' 라고 하며 딜러의 결과에 관계없이 무조건 건 금액을 잃게 된다.  
8) 참가자들은 블랙잭이 아닌 경우, 합계가 21점에 가까워지도록 하기 위해 딜러로부터 카드를 추가로 받을 수 있다(**Hit**). 추가 카드는 1장씩 몇 장이라도 요구할 수 있다. 예를 들면, K와 3을 받아 13점인 경우에 1장을 요구하여 그것이 2이면 15점이 되고, 다시 1장을 더 요구하여 그것이 6이면 합계 21점이 된다.  
9) 한편, 카드를 더 받지 않는 것이 유리하다고 판단되면 추가하지 않아도 된다(**Stand**). 예를 들면, Q와 9을 받아 19점인 경우에 1장을 더 추가하면 21점을 초과할 염려가 있기 때문에 그 상태에서 카드를 더 받지 않고 그칠 수 있다.  
10) 딜러의 점수와 비교해서 동점이면 무승부, 딜러보다 높으면 이기고 낮으면 지게 된다. 합계가 21점을 초과하면 0점(버스트, Burst)으로 간주되는데, 딜러가 21점을 초과한 경우도 0점이 되지만, 21점을 초과한 참가자의 베팅액을 딸 수 있다. 즉 둘 다 버스트면 딜러가 이긴다. <span style="color:red">(단, 본 수업의 미니프로젝트에서는 동점인 경우 딜러가 이기는 것으로 한다.)</span>  
11) 딜러는 참가자들의 추가가 모두 끝난 뒤에 자신이 카드를 추가할 것인지 여부를 결정한다. 정식 규칙에서 딜러는 가진 카드의 합계가 16점 이하이면 반드시 1장을 추가해야 하고, 17점 이상이면 추가할 수 없게 되어 있다.
