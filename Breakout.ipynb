```python
#okayteamletsgo
answer = True
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import math
import random
canvasheight = 400
canvaswidth = 400
collision = False
powerupmomentum = 0
nopowerup = True
powerballtimer = 500000
extendpadwidthtimer = 500000
#list of powerups
activatepowerup = False
poweruptype = None
powerup_position = [-10,500] 
powerupmomentum = 0
powerballOn = False
tripleballOn = False
def powerballOnfunction():
    global powerballOn, ballcolor
    powerballOn = True
    ballcolor = "Red"
def powerballOfffunction():
    global powerballOn, ballcolor
    powerballOn = False
    ballcolor = "White"
def extendpadwidthOnfunction():
    global extendpadwidthOn, firsttime
    extendpadwidthOn = True
    firsttime = True
def extendpadwidthOfffunction():
    global extendpadwidthOn, firsttime
    extendpadwidthOn = False
    firsttime = True
def tripleballOnfunction():
    global tripleballOn
    tripleballOn = True
def tripleballOfffunction():
    global tripleballOn
    tripleballOn = False
def starttripleballfunction():
    global ballslist
    ballslist.append(ball_position1)
    ballslist.append(ball_position2)
#initial ball dimensions
ball_position = [300,200]
radius = 10
ballvector = [-1,1]
ballspeed = 2
ballcolor = "White"
#pad dimensions
pad_width = 100
pad_height = 5
pad_position = [canvaswidth/2, canvasheight - pad_height/2.0]
pad_momentum = 0
pad_speed = 2

#brick dimensions
brick_width = 30
brick_thick = 10
#lives
life=3
#score
score=0
sound=simplegui.load_sound("http://www.soundjay.com/human/sounds/bone-crack-1.mp3")

#draw other balls function
ball_position1 = [300,200]
ball_position2 = [200,300]
ballvector1 = [.5, .5]
ballvector2 = [-.5, .5]
ballslist = [ball_position]
def getballvector(e):
    answer = None
    if e == ball_position1:
        answer = ballvector1
    
    elif e == ball_position2:
        answer = ballvector2
    else:
        answer = ballvector
    return answer
def drawballs(canvas):
    for q in ballslist:
        canvas.draw_circle(q, radius/2.0, radius, ballcolor)
        
        q[0] += ballspeed*getballvector(q)[0]
        q[1] += ballspeed*getballvector(q)[1]
        
        if q[0] < radius:
            getballvector(q)[0] *= -1
        if q[0] > canvaswidth - radius:
            getballvector(q)[0] *= -1
        if q[1] < 2*radius:
            getballvector(q)[1] *= -1
        
        if q[1] > (pad_position[1] - radius) and (pad_position[0] - q[0]) < pad_width/2.0:
            angle = (math.pi/2.0)*(q[0] - pad_position[0]) / (pad_width/2.0)
            getballvector(q)[0] = ballspeed*math.sin(angle)
            getballvector(q)[1] = -(ballspeed*math.cos(angle))
        

#activate power up?
def activatepowerupfunction():
    global activatepowerup, poweruptype
    
    if nopowerup == True:
        poweruptype = random.randint(1,4)
        if poweruptype == 1 or poweruptype == 2 or poweruptype == 3:
            print (poweruptype)
            activatepowerup = True
        else:
            activatepowerup = False
    else:
        activatepowerup = False
#test
powerballOfffunction()
extendpadwidthOfffunction()
firsttime = False
#brick list generator function (basically takes a first point, the adds bricks until x and y are too large)
def generatebricklist(firstbrickx,firstbricky):
    global brick_list
    brick_list = [[firstbrickx,firstbricky]]
    nextbrickx = firstbrickx+brick_width + 5
    nextbricky = firstbricky
    done = False
    while done == False:
      
        if nextbrickx < (canvaswidth - brick_width/2.0) and nextbricky < canvasheight/4.0:
           
            brick_list += [[nextbrickx, nextbricky]]
            nextbrickx += brick_width + 5
        elif nextbrickx > (canvaswidth - brick_width/2.0) and nextbricky < canvasheight/4.0:
            nextbrickx = firstbrickx
            nextbricky += brick_thick + 5
        else: 
            done = True



# brick list generator        
generatebricklist(((canvaswidth%(brick_width+5))+5)/2.0 + brick_width/2.0,50)
#dead_brick_list = [[200,200]]
dead_brick_list = []
#main (draw) function
def draw(canvas):
    global ball_position, radius, pad_momentum, score, life, powerup_position, powerupmomentum, nopowerup, powerballtimer, firsttime, pad_width, extendpadwidthtimer, sound, tripleballOn
    if extendpadwidthOn == True and firsttime == True:
        pad_width*=2
        firsttime = False
    if extendpadwidthOn == False and firsttime == True:
        pad_width*=0.5
        firsttime = False
    #paddle
    canvas.draw_line([pad_position[0]-pad_width/2.0,pad_position[1]], [pad_position[0]+pad_width/2.0,pad_position[1]], pad_height, "White")
    
    #ball
    
    #powerup
    if poweruptype == 1:
        canvas.draw_text("power ball!", powerup_position, 15, "Green")
    elif poweruptype ==2:
        canvas.draw_text("extend pad length", powerup_position, 15, "Purple")
    elif poweruptype ==3:
        canvas.draw_text("triple ball!", powerup_position, 15, "Orange")
    if powerballOn == True:
        canvas.draw_text("power ball timer = " + str(powerballtimer), [canvaswidth/2.0,20], 15, "Green")
    if extendpadwidthOn == True:
        canvas.draw_text("extend pad length timer = " + str(extendpadwidthtimer), [canvaswidth/2.0-50,20], 15, "Purple")
    #powerupmovement
    powerup_position[1] += powerupmomentum
    #paddlemovement
    pad_position[0] += pad_speed*pad_momentum
    
    #draw individual brick function
    def draw_brick(x,y):
        canvas.draw_line([x-brick_width/2.0, y], [x+brick_width/2.0, y], brick_thick, "White")
        
    def collisiontester(x,y):
        global score
        if abs(x[0]-s[0])<brick_width/2.0+radius+y[0] and abs(x[1]-s[1]) < brick_thick/2.0+radius+y[1]:
            collision = True
            activatepowerupfunction()
            score+=1            
            return collision
            
            
        else:
            collision = False
    def draw_dead_brick(x,y):
        canvas.draw_line([x-brick_width/2.0, y], [x+brick_width/2.0, y], brick_thick, "Red")
        
    for i in brick_list:
        draw_brick(i[0], i[1])

    for l in dead_brick_list:
        draw_dead_brick(l[0], l[1])
        
    #ball hit brick
    for s in brick_list:
        global activatepowerup
        for l in ballslist:
            if collisiontester(l, getballvector(l)) == True:
                if abs(l[0]-s[0])/brick_width < abs(l[1] - s[1])/brick_thick:
                    if powerballOn == False:
                        getballvector(l)[1] *= -1
                                    
                else:
                    if powerballOn == False:
                        getballvector(l)[0] *= -1
                sound.play()        
                brick_list.remove(s)
                
                l[0] += ballspeed*getballvector(l)[0]
                l[1] += ballspeed*getballvector(l)[1]
            if activatepowerup == True:
                powerup_position = [-10,500]
                powerup_position[0] = s[0]
                powerup_position[1] = s[1]
                powerupmomentum = 1
                activatepowerup = False
   
    canvas.draw_text("Score="+str(score), [0,20], 15, "white")        
    #brick hit dead brick
    for z in dead_brick_list:
        #ball from down
        if fabs(ball_position[1] - radius - (z[1] + brick_thick/2.0)) < fabs(ballvector[1]) and fabs(ball_position[0] - z[0]) < brick_width/2.0+2.5:
            ballvector[1] *= -1
        #ball from up
        elif fabs(ball_position[0] + radius - (z[1] - brick_width/2.0)) < fabs(ballvector[1]) and fabs(ball_position[0] - z[0]) < brick_thick/2.0+2.5:
            ballvector[1] *= -1
        #ball from left
        elif fabs(ball_position[0] + radius - (z[0] - brick_width/2.0)) < fabs(ballvector[0]) and fabs(ball_position[1] - z[1]) < brick_thick/2.0+2.5:
            ballvector[0] *= -1
            
        #ball from right
        elif fabs(ball_position[0] - radius - (z[0] + brick_width/2.0)) < fabs(ballvector[0]) and fabs(ball_position[1] - z[1]) < brick_thick/2.0+2.5:
            ballvector[0] *= -1
            
            
    global life
    canvas.draw_text("Lives="+str(life), [350,20], 15, "white")
    lives(canvas)
    
    #power up timers
    powerballtimer -= 1
    extendpadwidthtimer -=1
    if powerballtimer == 0:
        powerballOfffunction()
    if extendpadwidthtimer == 0:
        extendpadwidthOfffunction()
#other balls?
    drawballs(canvas)
#def end game    
    if life==0:
        canvas.draw_text("YOU LOSE! Press button to play again.",[0,250],25, "White")
    elif brick_list == []:
        canvas.draw_text("You win! Press button to play again.", [0,250],25,"White")
    if ball_position2[1] + radius >=canvasheight:
        ball_position2[1] = 10
        ballslist.pop(-1)
    if ball_position1[1] + radius >=canvasheight:
        if len(ballslist) == 2:
            ball_position1[1] = 10
            ballslist.pop(-1)
        elif len(ballslist) == 3:
            ball_position1[1] = 10
            ballslist.pop(1)
        if len(ballslist) == 1:
            tripleballOn = False
    powerups()
            
    
def lives(canvas):
    global life, ball_position,ballvector, nopowerup, powerballtimer, firsttime, extendpadwidthtimer
    if ball_position[1]+radius>=canvasheight:
        life-=1
        ballvector = [-1,1]
        ball_position = [200,100]
        ballslist.pop(0)
        ballslist.insert(0,ball_position)
        if life==0:
            global ballspeed
            ballspeed=0
            return ballspeed
def powerups():
    global life, ball_position,ballvector, nopowerup, powerballtimer, firsttime, extendpadwidthtimer
    if abs(powerup_position[0]-pad_position[0]) < (pad_width/2.0+15) and abs(powerup_position[1] - pad_position[1])<pad_height/2.0:
        if poweruptype == 1:
            powerballOnfunction()
            powerballtimer = 500
        elif poweruptype ==2:
            extendpadwidthOnfunction()
            extendpadwidthtimer = 500
        elif poweruptype ==3:
            tripleballOnfunction()
            starttripleballfunction()
        powerup_position[1] = canvasheight + 10
    if powerup_position[1] > canvasheight and powerballOn == False and extendpadwidthOn == False and tripleballOn == False:
        nopowerup = True
    else:
        nopowerup = False
    
#newgame
def new_game():
    global life, ball_position, pad_position, ballspeed, score, ballvector
    ball_position = [200,100]
    life=3
    ballspeed=2
    score=0
    pad_position = [canvaswidth/2, canvasheight - pad_height/2.0]
    ballvector = [-1,1]
    generatebricklist(((canvaswidth%(brick_width+5))+5)/2.0 + brick_width/2.0,50)   
    ballslist = [ball_position]
#keydown yo    
def keydown_handler(key):
    global pad_position, pad_momentum
    if key == 37:
        pad_momentum = -3
    if key == 39:
        pad_momentum = 3

#keyup oiy        
def keyup_handler(key):
    global pad_momentum
    if key == 37:
        pad_momentum = 0
    if key == 39:
        pad_momentum = 0

#all that ooey gooey frame stuff    
frame = simplegui.create_frame("What up dawg?", canvaswidth, canvasheight)
frame.set_draw_handler(draw)
frame.set_keydown_handler(keydown_handler)
frame.set_keyup_handler(keyup_handler)
frame.add_button("Play Again", new_game, 100)
frame.start()
```
