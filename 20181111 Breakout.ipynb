```python
# Implementation of classic arcade game Pong

import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import random
import math

WIDTH = 600
HEIGHT = 300       
BALL_RADIUS = 10
PAD_WIDTH = 8
PAD_HEIGHT = 80
HALF_PAD_WIDTH = PAD_WIDTH / 2
HALF_PAD_HEIGHT = PAD_HEIGHT / 2
ball_pos = [WIDTH / 2, HEIGHT / 2]
ball_vel = [0,0]
paddle_pos = (HEIGHT - PAD_HEIGHT)/2
paddle_vel = 0
life = 3
block_width = 15
block_height = 20

class Block():
    def __init__(self, color, x, y):
        super().__init__()
        self.image = pygame.Surface([block_width, block_height])
        self.image.fill(color)
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y

def spawn_ball():
    global ball_pos, ball_vel
    ball_pos = [WIDTH / 2, HEIGHT / 2]
    ball_vel = [3, -3]
        
# define event handlers
def new_game():
    global paddle_pos, paddle_vel
    global score 
    life = 3
    spawn_ball()
    
def draw(canvas):
    global life, paddle_pos, ball_pos, ball_vel, paddle_vel
    canvas.draw_line([WIDTH - PAD_WIDTH, 0],[WIDTH - PAD_WIDTH, HEIGHT], 1, "White")
        
    
    if ball_pos[0] <= BALL_RADIUS:
        ball_vel[0] = - ball_vel[0]
    if ball_pos[0] >= (WIDTH - BALL_RADIUS - PAD_WIDTH):
        if paddle_pos <= ball_pos[1] <= (paddle_pos+PAD_HEIGHT):
            ball_vel[0] = - 1.1 * ball_vel[0]
        else:
            life -= 1
            spawn_ball()
    if ball_pos[1] <= BALL_RADIUS:
        ball_vel[1] = - ball_vel[1]
    if ball_pos[1] >= (HEIGHT - BALL_RADIUS):
        ball_vel[1] = - ball_vel[1]
    
    # draw bal
    ball_pos[0] += ball_vel[0]
    ball_pos[1] += ball_vel[1]
    canvas.draw_circle(ball_pos, BALL_RADIUS, 0.1, "White", "White")
    
    if 0 <= (paddle_pos + paddle_vel) <= HEIGHT - PAD_HEIGHT:
        paddle_pos += paddle_vel
    
    paddle = canvas.draw_line([WIDTH - PAD_WIDTH/2, paddle_pos],[WIDTH - PAD_WIDTH/2, paddle_pos+PAD_HEIGHT], PAD_WIDTH, "White")
    
    canvas.draw_text("Lives="+str(life), [450,20], 15, "white")
    
def keydown(key):
    global paddle_vel
    vel = 3
    if key == simplegui.KEY_MAP["down"]:
        paddle_vel = vel
    if key == simplegui.KEY_MAP["up"]:
        paddle_vel = -vel
        
def keyup(key):
    global paddle_vel
    if key == simplegui.KEY_MAP["down"]:
        paddle_vel = 0
    if key == simplegui.KEY_MAP["up"]:
        paddle_vel = 0
    

frame = simplegui.create_frame("Pong", WIDTH, HEIGHT)
resstart = frame.add_button("Restart", new_game, 100)
frame.set_draw_handler(draw)
frame.set_keydown_handler(keydown)
frame.set_keyup_handler(keyup)


# start frame
new_game()
frame.start()
```
